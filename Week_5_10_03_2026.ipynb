```python
#Z-score
import numpy as np
data=[10,20,30,40,50]
mean=np.mean(data)
print("Mean", mean)
std_dev=np.std(data,ddof= 1)
print("Standard deviation",std_dev)
x_mean=40
z_score=(x_mean)/std_dev
print("Z-score of 40",z_score)
z_score=[(x_mean/std_dev for x in data)]
print("Z-score of data",z_score)
```

```python
#Statistical inference
import numpy as np
import math
import matplotlib.pyplot as plt
#population Dta
population=np.array([10,20,30,40,50,60,70])
N=len(population)
print("population size(N):",N)
mu=np.mean(population)
print("Population mean:",mu)
sigma=np.var(population,ddof=0)
print("Population variance:",sigma)
sigma=np.sqrt(sigma)
print("Population standard deviation:",sigma)
sample=np.array([20,30,40,50])
n=len(sample)
print("\n sample size(n):",n)
x_bar=np.mean(sample)
print("sample mean:",x_bar)
s2=np.var(sample,ddof=1)
print("sample variance(s2):",s2)
s=np.sqrt(s2)
print("sample standard deviation(s):",s)
SE=sigma/math.sqrt(n)
print("\n standard error:",SE)
z=(x_bar-mu)/SE
print("\n z-score:",z)
num_samples=1000
sample_size=4
sample_means=[]
for i in range(num_samples):
    samp=np.random.choice(population, sample_size,replace=True)
    sample_means.append(np.mean(samp))
sample_means=np.array(sample_means)
print("Mean of sample means:",np.mean(sample_means))
print("Std of sample means:",np.std(sample_means))
plt.hist(sample_means, bins=20,edgecolor='black')
plt.title("Central Limit Theorem: Distribution of Sample Means")
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")
plt.show()
```

```python
#Hypothesis Code
import math
n=100
x=60
p_hat=x/n
SE=math.sqrt(p_hat*(1-p_hat)/n)
z=1.96
lower=p_hat-z*SE
upper=p_hat+z*SE
print("Sample proportion=",p_hat)
print("Confidence Interval=",(lower,upper))
p0=0.50
z_test=(p_hat-p0)/math.sqrt(p0*(1-p0)/n)
print("z value=",z_test)
if abs(z_test)>1.96:
  print("Reject Ho")
else:
  print("Do not reject Ho")
```
