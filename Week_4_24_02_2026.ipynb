```python
#Descriptive statistic functions
import pandas as pd
data={
    "Ages":[23,34,25,28,36,40,29,42,32,45],
    "Salaries":[3000,3200,5500,4000,5200,6000,3500,3700,4500,4800]
}
df = pd.DataFrame(data)
print("Summary statistics:\n",df.describe())
print("Mean",df.mean())
print("Median",df.median())
print("Mode",df.mode())
print("Standard deviation",df.std())
print("Variance",df.var())
print("Count",df.count())
print("Skewness",df.skew())
print("Kurtosis",df.kurt())
```

```python
#Data Distribution
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
normal_data=np.random.normal(loc=50,scale=6,size=100)
right_skewed=np.random.exponential(scale=3,size=100)
left_skewed=-np.random.exponential(scale=5,size=100)
plt.hist(normal_data,bins=40)
plt.title("Normal Distribution")
plt.show()
plt.hist(right_skewed,bins=40)
plt.title("Right Skewed Distribution")
plt.show()
plt.hist(left_skewed,bins=40)
plt.title("Left Skewed Distribution")
plt.show()
print("Normal Data Mean:",np.mean(normal_data))
print("Normal Data Median:",np.median(normal_data))
print("Right Skewed Mean:",np.mean(right_skewed))
print("Right Skewed Median:",np.median(right_skewed))
print("Left Skewed Mean:",np.mean(left_skewed))
print("Left Skewed Median:",np.median(left_skewed))
```

```python
#Data Analysis
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
data=[15,16,18,20,22,23,25,26,27,29,30,31,32,33,35,38,41,42,45]
#Mean & Median
mean_val=np.mean(data)
median_val=np.median(data)
print("Mean of the data:",mean_val)
print("Median of the data:",median_val)
# Skewness
if mean_val > median_val:
    print("Graph is right-skewed")
elif mean_val < median_val:
    print("Data is slightly left-skewed")
else:
    print("Data is symmetric")
```

```python
#Create DataFrame using Bmi
data_dict={
    "BMI":[15,16,18,20,22,23,25,26,27,29,30,31,32,33,35,38,41,42,45]
}
df=pd.DataFrame(data_dict)
#Plot Histogram
plt.hist(df["BMI"],bins=6)
plt.title("BMI Data Distribution")
plt.xlabel("BMI Values")
plt.ylabel("Frequency")
plt.show()
print("Mean:",df["BMI"].mean())
print("Median:",df["BMI"].median())
print("Skewness:",df["BMI"].skew())
```

```python
#Create a dataset with missing values
import pandas as pd
data={
    'Name':['Alice','Bob','Sara','David','Eve'],
    'Age':[26,None,32,42,29],
    'Salary':[50000, 60000, None, 70000, 61000],
    'Allowances':[2000, None, 2000, 2000, 2500]
}
df=pd.DataFrame(data)
print("Missing Values Count:\n",df.isnull().sum())
df_cleaned=df.dropna()
print("\nDataFrame after dropping missing values:\n",df_cleaned)
df_filled=df.fillna({'Age':df['Age'].mean(),'Salary':df['Salary'].mean(),'Allowances':2000})
print("\nDataFrame after filling missing values:\n",df_filled)
```

```python
#IQR method
import pandas as pd
data={
    "name":['A','B','C','D','E','F'],
    "age":[12,22,26,110,25,23]
}
df=pd.DataFrame(data)
q1=df['age'].quantile(0.25)
q3=df['age'].quantile(0.75)
iqr=q3-q1
lower_bound=q1-1.5*iqr
upper_bound=q3+1.5*iqr
outliners=df[(df['age']<lower_bound)|(df['age']>upper_bound)]
print("outliners using IQR method")
print(outliners)
```
